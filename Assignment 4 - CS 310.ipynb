```python
# data preparation, get 6 stocks data from Excel files
aapl = pd.ExcelFile('AAPL.xlsx')
amzn = pd.ExcelFile('AMZN.xlsx')
fb = pd.ExcelFile('FB.xlsx')
goog = pd.ExcelFile('GOOG.xlsx')
ibm = pd.ExcelFile('IBM.xlsx')
msft = pd.ExcelFile('MSFT.xlsx')
```

```python
print(aapl.sheet_names)
```

```python
# pick sheet from Excel file and load the data into dataframes using Pandas
df_aapl = aapl.parse('AAPL')
df_amzn = amzn.parse('AMZN')
df_fb = fb.parse('FB')
df_goog = goog.parse('GOOG')
df_ibm = ibm.parse('IBM')
df_msft = msft.parse('MSFT')
```

```python
# delete unnecessary columns for this assignment from dataframes
del df_aapl['Open']
del df_aapl['High']
del df_aapl['Low']
del df_aapl['Volume']
del df_amzn['Open']
del df_amzn['High']
del df_amzn['Low']
del df_amzn['Volume']
del df_fb['Open']
del df_fb['High']
del df_fb['Low']
del df_fb['Volume']
del df_goog['Open']
del df_goog['High']
del df_goog['Low']
del df_goog['Volume']
del df_ibm['Open']
del df_ibm['High']
del df_ibm['Low']
del df_ibm['Volume']
del df_msft['Open']
del df_msft['High']
del df_msft['Low']
del df_msft['Volume']
```

```python
print(df_aapl)
```

```python
# create columns with ratio of previous day's Close price to today's Close price
a = 'prev day Close price / today Close price'
df_aapl[a] = df_aapl['Close'].shift(1)/df_aapl['Close']
df_amzn[a] = df_amzn['Close'].shift(1)/df_amzn['Close']
df_fb[a] = df_fb['Close'].shift(1)/df_fb['Close']
df_goog[a] = df_goog['Close'].shift(1)/df_goog['Close']
df_ibm[a] = df_ibm['Close'].shift(1)/df_ibm['Close']
df_msft[a] = df_msft['Close'].shift(1)/df_msft['Close']
```

```python
print(df_aapl)
```

```python
# create columns with ratio of previous day's Adj Close price to today's Adj Close price
b = 'prev day Adj Close price / today Adj Close price'
df_aapl[b] = df_aapl['Adj Close'].shift(1)/df_aapl['Adj Close']
df_amzn[b] = df_amzn['Adj Close'].shift(1)/df_amzn['Adj Close']
df_fb[b] = df_fb['Adj Close'].shift(1)/df_fb['Adj Close']
df_goog[b] = df_goog['Adj Close'].shift(1)/df_goog['Adj Close']
df_ibm[b] = df_ibm['Adj Close'].shift(1)/df_ibm['Adj Close']
df_msft[b] = df_msft['Adj Close'].shift(1)/df_msft['Adj Close']
```

```python
print(df_aapl)
```

```python
# create columns with difference between (the ratio of previous day's Adj Close price to today's Adj Close price)
# and (the ratio of previous day's Close price to today's Close price)
c = '(prev day Adj Close price / today Adj Close price) - (prev day Close price / today Close price)'
df_aapl[c] = df_aapl[b] - df_aapl[a]
df_amzn[c] = df_amzn[b] - df_amzn[a]
df_fb[c] = df_fb[b] - df_fb[a]
df_goog[c] = df_goog[b] - df_goog[a]
df_ibm[c] = df_ibm[b] - df_ibm[a]
df_msft[c] = df_msft[b] - df_msft[a]
```

```python
print(df_aapl)
```

```python
# create column of Dividend by taking the difference found in column C and multiplying it by the Close price column
d = 'Dividend'
df_aapl[d] = df_aapl[c] * df_aapl['Close']
df_amzn[d] = df_amzn[c] * df_amzn['Close']
df_fb[d] = df_fb[c] * df_fb['Close']
df_goog[d] = df_goog[c] * df_goog['Close']
df_ibm[d] = df_ibm[c] * df_ibm['Close']
df_msft[d] = df_msft[c] * df_msft['Close']
```

```python
print(df_aapl)
```

```python
# delete unnecessary columns for the final dataframes
del df_aapl['Close']
del df_aapl['Adj Close']
del df_aapl[a]
del df_aapl[b]
del df_aapl[c]
del df_amzn['Close']
del df_amzn['Adj Close']
del df_amzn[a]
del df_amzn[b]
del df_amzn[c]
del df_fb['Close']
del df_fb['Adj Close']
del df_fb[a]
del df_fb[b]
del df_fb[c]
del df_goog['Close']
del df_goog['Adj Close']
del df_goog[a]
del df_goog[b]
del df_goog[c]
del df_ibm['Close']
del df_ibm['Adj Close']
del df_ibm[a]
del df_ibm[b]
del df_ibm[c]
del df_msft['Close']
del df_msft['Adj Close']
del df_msft[a]
del df_msft[b]
del df_msft[c]
```

```python
print(df_aapl)
```

```python
# change Null values in the Dividend column to values of 0
df_aapl['Dividend'] = df_aapl['Dividend'].fillna(0)
df_amzn['Dividend'] = df_amzn['Dividend'].fillna(0)
df_fb['Dividend'] = df_fb['Dividend'].fillna(0)
df_goog['Dividend'] = df_goog['Dividend'].fillna(0)
df_ibm['Dividend'] = df_ibm['Dividend'].fillna(0)
df_msft['Dividend'] = df_msft['Dividend'].fillna(0)

# round Dividend values to 2 decimal places
df_aapl = df_aapl.round({"Dividend":2})
df_amzn = df_amzn.round({"Dividend":2})
df_fb = df_fb.round({"Dividend":2})
df_goog = df_goog.round({"Dividend":2})
df_ibm = df_ibm.round({"Dividend":2})
df_msft = df_msft.round({"Dividend":2})

# filter out and delete rows for which Dividend values are zero
df_aapl = df_aapl[df_aapl['Dividend'] != 0]
df_amzn = df_amzn[df_amzn['Dividend'] != 0]
df_fb = df_fb[df_fb['Dividend'] != 0]
df_goog = df_goog[df_goog['Dividend'] != 0]
df_ibm = df_ibm[df_ibm['Dividend'] != 0]
df_msft = df_msft[df_msft['Dividend'] != 0]

# remove index column by setting index column to Date instead
df_aapl = df_aapl.set_index('Date')
df_amzn = df_amzn.set_index('Date')
df_fb = df_fb.set_index('Date')
df_goog = df_goog.set_index('Date')
df_ibm = df_ibm.set_index('Date')
df_msft = df_msft.set_index('Date')

# get positive values of Dividend
df_aapl = -df_aapl
df_amzn = -df_amzn
df_fb = -df_fb
df_goog = -df_goog
df_ibm = -df_ibm
df_msft = -df_msft
```
